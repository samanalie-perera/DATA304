### Problem 3: Consider the distribution given by a two-component mixture of Normal random variables with p = (p, 1 − p), μ1 = μ2 = 0, σ2 1 = 1, σ2 2 = 10. Plot several of the resulting densities choosing different values of p.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import expon, norm

#generate three exponential density curves, that have different rates. 
x = np.linspace(0, 5, 1000)
y1 = expon.pdf(x, scale=1)
plt.plot(x, y1, label='Rate=1')

y2 = expon.pdf(x, scale=0.5)
plt.plot(x, y2, label='Rate=0.5')

y3 = expon.pdf(x, scale=2)
plt.plot(x, y3, label='Rate=2')

plt.xlim(0, 5)
plt.yscale('log')
plt.legend()
plt.show()


#loc is mean
#scale is standard deviation
x = np.linspace(-3, 3, 1000)

y4 = norm.pdf(x, loc=0, scale=1)
plt.plot(x, y4, label='mean=0, sd=1')

y5 = 0.5*norm.pdf(x, loc=0, scale=1) + 0.5*norm.pdf(x, loc=-1, scale=1)
plt.plot(x, y5, label='Mixture')


plt.xlim(-3, 3)
plt.yscale('log')
plt.legend()
plt.show()

d1 = norm.pdf(x, loc=0, scale=1)
d2 = 0.5*norm.pdf(x, loc=0, scale=1) + 0.5*norm.pdf(x, loc=-1, scale=1)
data = pd.DataFrame({'x': x, 'd1': d1, 'd2': d2})

plt.plot(data['x'], data['d1'], color='red', label='mean=0, sd=1')
plt.plot(data['x'], data['d2'], label='Mixture')
plt.legend()
plt.show()
```

Visually, using a linear scale, with which values of p you are unable to notice that the mixture
is not a Normal density?

```python
ps = np.linspace(0, 1, 101)
results = []

for p in ps:
    x = np.linspace(-5, 5, 1000)
    y = p * norm.pdf(x, loc=0, scale=1) + (1 - p) * norm.pdf(x, loc=-1, scale=1)
    y_normal = norm.pdf(x, loc=0, scale=1)
    l2_normal = np.sqrt(np.sum((y - y_normal)**2))
    results.append({'p': p, 'l2_normal': l2_normal})
    
resultsDF = pd.DataFrame(results)
plt.plot(resultsDF['p'], resultsDF['l2_normal'])
plt.xlabel('p')
plt.ylabel('L2 normal')
plt.show()
```

Visually, using a semilogarithmic scale, with which values of p you are unable to notice that the mixture is not a Normal density?

```python
ps = np.linspace(0, 1, 101)
results = []

for p in ps:
    x = np.linspace(-5, 5, 1000)
    y = p * norm.pdf(x, loc=0, scale=1) + (1 - p) * norm.pdf(x, loc=-1, scale=1)
    y_normal = norm.pdf(x, loc=0, scale=1)
    l2_normal = np.sqrt(np.sum((y - y_normal)**2))
    results.append({'p': p, 'l2_norm': l2_normal})
    
resultsDF = pd.DataFrame(results)
plt.semilogy(resultsDF['p'], resultsDF['l2_norm'])
plt.xlabel('p')
plt.ylabel('L2 norm')
plt.show()
```
